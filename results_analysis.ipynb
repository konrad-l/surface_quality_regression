```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Results

```python
df = pd.read_csv('results/results_2024_04_14_17_12.csv')
df
```

```python
df['train_r2'] = df['train_r2']*100
df['test_r2'] = df['test_r2']*100
df
```

```python
ra_df = df[df['target'] == 'Ra']
ra_df = ra_df.sort_values(by=['test_r2'], ascending=False)
ra_df
```

```python
wz_df = df[df['target'] == 'Wz']
wz_df = wz_df.sort_values(by=['test_r2'], ascending=False)
wz_df
```

# Plot Metrics

```python
targets = df.target.unique()

def plot_metrics(target):
    df_new = df[df['target'] == target]
    df_long = (
        df_new[
            ['model_name','train_r2','train_rmse','test_r2','test_rmse']]
            .melt(id_vars=['model_name'], var_name='Metric', value_name='Score')
    )
    
    plt.figure(figsize=(12, 5))
    sns.barplot(data=df_long, x='Metric', y='Score', hue='model_name')
    plt.suptitle(f"Target: {target}")
    plt.title(f'Validation Scores by Model')
    plt.xticks(rotation=45)
    plt.ylabel('Score')
    plt.xlabel('Validation Metric')
    plt.legend(title='Model Name', loc='upper right')
    plt.show()

for target in targets:
    plot_metrics(target=target)
```
